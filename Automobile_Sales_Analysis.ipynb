## RFM Customer Segmentation

## RFM  Analysis

### recency

```python
recency= df_clean.groupby(["CUSTOMERNAME"])["ORDERDATE"].max().reset_index()
recency["Days_science_lastorder"] = (df_clean["ORDERDATE"].max()-recency["ORDERDATE"]).dt.days
recency.drop(columns="ORDERDATE",inplace=True)
recency.sort_values(by="Days_science_lastorder",ascending=False,inplace=True,)
recency
```

```python
R = df_clean.groupby(["CUSTOMERNAME","ORDERDATE"])["SALES"].sum().reset_index().sort_values(by=["CUSTOMERNAME","ORDERDATE"])

R["avg_purchase_interval"] = R.groupby("CUSTOMERNAME")["ORDERDATE"].diff().dt.days

R.fillna(0,inplace=True)

R = R.groupby("CUSTOMERNAME")["avg_purchase_interval"].mean().reset_index().round()
R
```

## Frequency

```python
F = df.groupby(["CUSTOMERNAME","ORDERNUMBER"])["SALES"].sum().reset_index()
F = F.groupby("CUSTOMERNAME").agg({"ORDERNUMBER":"count","SALES":"sum"}).reset_index()
F["AVG ORDER VALUE"] = (F["SALES"]/F["ORDERNUMBER"]).round()
F.rename(columns={"ORDERNUMBER":"Total Orders"},inplace=True)
F.drop(columns="SALES",inplace=True)
F
```

* Most of our highest-value customers, in terms of revenue, are the most active, and their spending patterns per order fall into two categories: either high activity with low to medium spending, or lower activity with high spending.

```python
M= df.groupby(["CUSTOMERNAME"])["SALES"].sum().reset_index()
M
```

###

### RFM Customer Segmentation & Churn Risk

```python
RFM =pd.merge(R,F,on="CUSTOMERNAME")
RFM =pd.merge(RFM,M,on="CUSTOMERNAME").sort_values(by="SALES",ascending=False)


RFM["DAYS_SINCE_LASTORDER"] = recency["Days_science_lastorder"]
RFM.rename(columns={"SALES":"MONETARY","ORDERNUMBER":"Frequency"},inplace=True)
RFM["Customers Statue"] = pd.qcut(RFM["MONETARY"],q=3,labels=["standard","Loyal","VIP"])


RFM['potential churn risk'] = 'At Risk'

RFM.loc[RFM['avg_purchase_interval'] == 0,'potential churn risk'] = 'One-Time Customer'

RFM.loc[
    (RFM['avg_purchase_interval'] > 0) &
    (RFM['DAYS_SINCE_LASTORDER'] <= RFM['avg_purchase_interval']),
    'potential churn risk'
] = 'Regular'


# Potential churn customers
RFM.loc[
    (RFM['avg_purchase_interval'] > 0) &
    (RFM['DAYS_SINCE_LASTORDER'] >= RFM['avg_purchase_interval'] * 2),
    'potential churn risk'
] = 'Potential Churn'
```

```python
pct_churn = (RFM[RFM["potential churn risk"]=="Potential Churn"]["potential churn risk"].count()/RFM.shape[0])*100
pct_one_time = (RFM[RFM["potential churn risk"]=="One-Time Customer"]["potential churn risk"].count()/RFM.shape[0])*100
pct_regular = (RFM[RFM["potential churn risk"]=="Regular"]["potential churn risk"].count()/RFM.shape[0])*100
pct_at_risk = (RFM[RFM["potential churn risk"]=="At Risk"]["potential churn risk"].count()/RFM.shape[0])*100
print(f"The Percentage of  churned One Time Customers: {round(pct_one_time,2)}%")

print(f"The Percentage of Regular Customers: {round(pct_regular,2)}%")
print(f"The Percentage of May be churned Customers: {round(pct_churn,2)}%")
print(f"The Percentage of  Customers at risk: {round(pct_at_risk,2)}%")
VIP_risk = RFM[(RFM["potential churn risk"]=="Potential Churn")&(RFM["Customers Statue"]== "VIP")]["Customers Statue"].count()/RFM.shape[0]*100
VIP_risk
```

```python
revenue_distribution  = RFM.groupby('Customers Statue')["MONETARY"].agg(["sum","count"]).reset_index()
revenue_distribution = revenue_distribution.rename(columns={"sum":"Monetary","count":"Customer number"})

revenue_distribution["Percentage of Revenue"] = (revenue_distribution["Monetary"]/company_revenue*100).round()
revenue_distribution.sort_values(by="Monetary",ascending=False)
```

# Geographic sales analysis over time

```python
country_revenue = df_clean.groupby(["COUNTRY","ORDERDATE"])["SALES"].sum().reset_index(name="Revenue")
country_revenue["ORDERDATE"] = country_revenue["ORDERDATE"].dt.year
country_revenue = country_revenue.pivot_table(index="COUNTRY",columns="ORDERDATE",values="Revenue",aggfunc="sum")
country_revenue["2019 Growth"] = ((country_revenue[2019]-country_revenue[2018])/country_revenue[2018]*100).round()
country_revenue.fillna(0,inplace=True)

country_revenue["total revenue"] = country_revenue.sum(axis=1)
country_revenue["total revenue percentage"] = round((country_revenue["total revenue"]/company_revenue)*100)
country_revenue.sort_values(by="total revenue",ascending=False,inplace=True)
country_revenue.reset_index(inplace=True)

country_revenue.columns.name = None
country_revenue
```

```python
bins = [-float('inf'), -1,0, 0.05, 0.15, float('inf')]
labels = ['Declining',"New entry", 'Low Growth', 'Stable Growth', 'High Growth']

country_revenue['Market_Priority'] = pd.cut(country_revenue['2019 Growth'], bins=bins, labels=labels)
country_revenue.sort_values(by="2019 Growth", ascending=False,inplace=True)
country_revenue

pec_of_growth_country = country_revenue[country_revenue["2019 Growth"]>15].shape[0]/country_revenue.shape[0]*100
pec_of_Declining_country = country_revenue[country_revenue["2019 Growth"]<0].shape[0]/country_revenue.shape[0]*100
pec_of_new_entry = country_revenue[country_revenue["2019 Growth"]==0].shape[0]/country_revenue.shape[0]*100
pec_of_Declining_country = country_revenue[country_revenue["2019 Growth"]<=-1].shape[0]/country_revenue.shape[0]*100
print(f"The Percentage of Countries with Low Growth: {round(pec_of_Declining_country)}%")
print(f"The Percentage of Countries with High Growth: {round(pec_of_growth_country)}%")
print(f"The Percentage of Countries with New Entry: {round(pec_of_new_entry)}%")
print(f"The Percentage of Countries with Declining Growth: {round(pec_of_Declining_country)}%")
country_revenue

```
